```python
'''-----------------------------------------------------------------------------------------------------------------------------------------------------------------'''
def get_Fresh_itemset ():
    sorted_itemsets = ""                  #String Variable After Sorting and Fresh The Itemset
    itemset = ""                         # String Variable to get only the Letters

    for i in dic:                         #Looping on the dic to remove anything such Letters
        if i != "[" and i !="]" and i != "\"" and i != ","  and i != "=" and i != " " and i !="(" and i !=")" and i !="{" and i !="}":
            itemset += i

    #Our itemset is now ready to use - itemsets - but need to remove the repeated Letters

    sorted_itemsets = ''.join(sorted(set(itemset) , key = str.lower)).strip() #Sorting from A -> Z , Set ( Remove repetitions ) , Strip ( to remove Spaces )
    return sorted_itemsets

'''-----------------------------------------------------------------------------------------------------------------------------------------------------------------'''
'''-----------------------------------------------------------------------------------------------------------------------------------------------------------------'''

def Calc_Support_Of_FItemset (sorted_itemsets):
    dictionary_itemset = {}  # Use Dictionary to Store itemset with its Support

    for i in sorted_itemsets:
        letter = i
        support = 0

        dictionary_itemset[letter] = support


    temp_Dic = []                         # temp Dictionary To Use Temporal algorithm
    for i in dic:
        if i == "]":
            temp_Dic = set(temp_Dic)
            for j in temp_Dic:
                if j in dictionary_itemset.keys():
                    dictionary_itemset[j] += 1
            temp_Dic = []
        else:
            temp_Dic += i

    return dictionary_itemset
# Our itemset is Ready To use with its Support ( dictionary_itemset )
'''-----------------------------------------------------------------------------------------------------------------------------------------------------------------'''
'''-----------------------------------------------------------------------------------------------------------------------------------------------------------------'''

def Get_Mininum_Support():
    min_sup = input("Please Enter The Minimum Support :  ")

    return min_sup
'''-----------------------------------------------------------------------------------------------------------------------------------------------------------------'''
'''-----------------------------------------------------------------------------------------------------------------------------------------------------------------'''

def Get_FFrequent_Itemset(dictionary_itemset , min_supp):

    for i in dictionary_itemset.copy():

        if dictionary_itemset[i] < min_supp:

            del dictionary_itemset[i]

    return dictionary_itemset
'''-----------------------------------------------------------------------------------------------------------------------------------------------------------------'''
def New_Sorted_Itemset(sorted_itemset , dictionary_itemset):

    for i in dictionary_itemset.keys():
        sorted_itemset = dictionary_itemset.keys()

    return sorted_itemset

'''-----------------------------------------------------------------------------------------------------------------------------------------------------------------'''

def Temporal_Join(sorted_itemset):

    temporal_str = ""
    temporal_dic = {}
    str = ""

    for i in sorted_itemset:
        for j in sorted_itemset:
            temporal_str += i + j + ","

    for i in temporal_str:
        if i != ",":
           str += i
        else:
            temporal_dic[str] = 0
            str = ""

    return temporal_dic
'''-----------------------------------------------------------------------------------------------------------------------------------------------------------------'''

def Calc_Support_Of_SItemset(temporal_join_set , itemset):
    temp_dic = ""
    tmp =""
    for i in itemset:
        if i != "]":
            temp_dic += i
        else:
            for j in temp_dic:
                temp_dic = temp_dic.replace(j , "" ,1)
                tmp = temp_dic
                for u in tmp:#AB{FG}CD -- #B{FG}CD
                    if j+u in temporal_join_set.keys():
                        try:
                            temporal_join_set[j+u] += 1
                        except:
                            print("Not Frequent")
            temp_dic = ""
            tmp = ""
    temp_dic = ""
    txt = ""
    for i in itemset:
        if i != "]":
            temp_dic += i
        else:
            try:
                start = temp_dic.index("{")
                end = temp_dic.index("}")

                for u in range(start + 1, end):
                    txt += temp_dic[u]
                    if len(txt) == 2:
                        temporal_join_set[txt] -= 1
                temp_dic = ""
                txt = ""
            except:
                print("NONE Brackets")  
    return temporal_join_set

'''-----------------------------------------------------------------------------------------------------------------------------------------------------------------'''



'''  ______ ------***    MAIN FUNCTION    ***------______  '''

# Get The Dataset fron the file "itemset.txt."
inp = open("itemset.txt", "r")  # Open The file
dic = inp.read()  # Put the dataset in dic
#---------------------------------------------------------------------------------------

itemset = ""  # String Variable to get only the Letters

for i in dic:  # Looping on the dic to remove anything such Letters
    if i != "["  and i != "\"" and i != "," and i != "=" and i != " " and i != "(" and i != ")" :
        itemset += i

print("  -----  " + itemset)
#---------------------------------------------------------------------------------------
SORTED_ITEMSET                   = get_Fresh_itemset()
DICTIONARY_FITEMSET              = Calc_Support_Of_FItemset(SORTED_ITEMSET)
MIN_SUPPORT                      = Get_Mininum_Support()
FIRST_FREQ_ITEMSET               = Get_FFrequent_Itemset(DICTIONARY_FITEMSET , int(MIN_SUPPORT))
NEW_SORTED_ITEMS                 = New_Sorted_Itemset(SORTED_ITEMSET , DICTIONARY_FITEMSET)
TEMPORAL_JOIN_SET                = Temporal_Join(NEW_SORTED_ITEMS)
DICTIONARY_SITEMSET              = Calc_Support_Of_SItemset(TEMPORAL_JOIN_SET , itemset)
SECOND_FREQ_ITEMSET              = Get_FFrequent_Itemset(DICTIONARY_SITEMSET , int(MIN_SUPPORT))
SECOND_FREQ_ITEMSETnontemp       = NonTempJoin()
SECOND_FREQ_ITEMSETnontempsup    = Calc_Supp_Of_2Itemset(SECOND_FREQ_ITEMSETnontemp , itemset)
THREE_ITEMSETNON                 = join3ItemNONTemp(SECOND_FREQ_ITEMSETnontempsup,SECOND_FREQ_ITEMSET)
THREE_ITEMSET                    = Join3Item(SECOND_FREQ_ITEMSET)
FOUR_ITEMSET                     =Join4Item(THREE_ITEMSET)


#print(SORTED_ITEMSET)
#print(DICTIONARY_FITEMSET)
#print(MIN_SUPPORT)
#print(FIRST_FREQ_ITEMSET)
#print(NEW_SORTED_ITEMS)
#print(DICTIONARY_SITEMSET)
#print(TEMPORAL_JOIN_SET.keys())
#print(SECOND_FREQ_ITEMSET)
print("Frequent 1 itemsets :",DICTIONARY_FITEMSET)
print("Frequent 2 Temporal itemsets :",SECOND_FREQ_ITEMSET)
print("Frequent 2 Non-Temporal itemsets :",SECOND_FREQ_ITEMSETnontempsup)
print("3 Non-Temporal itemsets ",THREE_ITEMSETNON)
print("3 Temporal itemsets ",THREE_ITEMSET)
print("4 itemsets ",FOUR_ITEMSET)
#print(SECOND_FREQ_ITEMSETnontempsup)
inp.close()
```
